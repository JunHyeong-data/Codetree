```python
import numpy as np
import matplotlib.pyplot as plt
heights = [183, 176, 169, 186, 177, 173, 160, 185, 175, 155] # cm
weights = [86, 74, 59, 95, 80, 68, 50, 99, 85, 45] # kg

np_heights = np.array(heights)
np_weights = np.array(weights)

np_heights = np_heights/100

BMI = np_weights / np_heights**2
plt.hist(BMI[BMI>=25])
plt.show()

print(" BMI :", np.mean(BMI))
```

```python
import numpy as np
import matplotlib.pyplot as plt

nor_dist = np.random.normal(50,5,1000)
outlier = (nor_dist > 60)|(nor_dist < 40)

tot_outlier = np.sum(outlier)
pro_outlier = tot_outlier/len(nor_dist)

plt.hist(nor_dist[outlier],bins=20)
plt.title("Outlier Histogram")

print(" :", tot_outlier)
print(" :", pro_outlier)
```

```python
import numpy as np
np.random.seed(0)
A = np.random.randint(0, 10, size=(6, 6))

arr = np.zeros((3,3), dtype=int)

for i in range(3):
  for j in range(3):
    block = A[i*2:i*2+2, j*2:j*2+2]
    arr[i,j] = np.sum(block)

print(arr)
```

```python
import numpy as np
np.random.seed(0)
A = np.random.randint(1, 10, size=(20, 20))

A[A <= np.mean(A)] = 0

count = np.sum(A != 0)
print("0   ", count)
print("", np.mean(A[A != 0]))
```

```python
import numpy as np

#   (0~9    100*100 )
np.random.seed(0)
A = np.random.randint(0, 10, size=(100, 100))

k = 5
max_sum = -1
max_pos = (None, None)

for i in range(100):
    for j in range(100 - k + 1):
        in_sum = np.sum(A[i, j:j+k])
        if in_sum > max_sum:
            max_sum = in_sum
            max_pos = (i, j)

print("    :", max_sum)
print(":", max_pos)
```
