#실습 : 소수 찾기

```python
N_PRIMES = 50 #찾아야 하는 소수의 개수
number = 2 # 2부터 시작한다.
count = 0

while count < N_PRIMES:
  for i in range(2, int((number ** 0.5) + 1)):
    if number % i == 0: # 소수가 아님
      break
  else: # 소수
    count += 1
    print(number, end=' ')
  number += 1
```
