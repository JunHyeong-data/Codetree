```python
print(":")
print()
if (x := float(input("x  : "))) <= 0:
    func1 = x**2 - 9*x + 2
    print(f"f(x)  {func1:.4f}")
else:
    func2 = 7*x + 2
    print(f"f(x)  {func2:.4f}")
```

# def  

```python
print(":")
print()
x = float(input("x  : "))
def func(x) :
  if x <= 0 :
    func1 = x**2 - 9*x + 2
    return func1
  else :
    func2 = 7*x + 2
    return func2
print(f"f(x)  {func(x):.4f}")
```
