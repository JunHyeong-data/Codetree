# 바다코끼리 연산자 사용

```python
print("실행결과:")
print()
if (x := float(input("x의 값을 입력하시오: "))) <= 0:
    func1 = x**2 - 9*x + 2
    print(f"f(x)의 값은 {func1:.4f}")
else:
    func2 = 7*x + 2
    print(f"f(x)의 값은 {func2:.4f}")
```

# def 함수 사용

```python
print("실행결과:")
print()
x = float(input("x의 값을 입력하시오: "))
def func(x) :
  if x <= 0 :
    func1 = x**2 - 9*x + 2
    return func1
  else :
    func2 = 7*x + 2
    return func2
print(f"f(x)의 값은 {func(x):.4f}")
```
